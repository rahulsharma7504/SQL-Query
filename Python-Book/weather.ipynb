```python
import matplotlib.pyplot as plt
from matplotlib.widgets import CheckButtons
import requests

# API key
api_key = "bf5dfdfae8848c9763983934045d2bb3"

# City name
city = "Agra"

# Fetch weather data for today, tomorrow, and the day after tomorrow
response_today = requests.get(
    f"https://api.openweathermap.org/data/2.5/forecast?q={city}&appid={api_key}&units=metric"
).json()["list"]
response_tomorrow = requests.get(
    f"https://api.openweathermap.org/data/2.5/forecast?q={city}&appid={api_key}&units=metric&cnt=2&dt=1641824000"
).json()["list"]
response_day_after_tomorrow = requests.get(
    f"https://api.openweathermap.org/data/2.5/forecast?q={city}&appid={api_key}&units=metric&cnt=2&dt=1641910400"
).json()["list"]

# Check if the lists are not empty before accessing elements
if response_today and response_tomorrow and response_day_after_tomorrow:
    # Get the time and temperature data
    today_time = [entry["dt"] for entry in response_today]
    today_temperature = [entry["main"]["temp"] for entry in response_today]
    tomorrow_time = [entry["dt"] for entry in response_tomorrow]
    tomorrow_temperature = [entry["main"]["temp"] for entry in response_tomorrow]
    day_after_tomorrow_time = [entry["dt"] for entry in response_day_after_tomorrow]
    day_after_tomorrow_temperature = [entry["main"]["temp"] for entry in response_day_after_tomorrow]

    # Create the figure and axes
    fig, ax = plt.subplots()

    # Plot the line graph
    line_today, = ax.plot(today_time, today_temperature, label="Aaj", marker='o')
    line_tomorrow, = ax.plot(tomorrow_time, tomorrow_temperature, label="Kal", marker='o')
    line_day_after_tomorrow, = ax.plot(day_after_tomorrow_time, day_after_tomorrow_temperature, label="Parson", marker='o')

    # Add a title and labels
    ax.set_title("Weather data for Agra")
    ax.set_xlabel("Time")
    ax.set_ylabel("Temperature (°C)")

    # Add the checkboxes
    lines = [line_today, line_tomorrow, line_day_after_tomorrow]
    check_buttons = CheckButtons(ax, ['Aaj', 'Kal', 'Parson'], [True, True, True])

    def toggle_visibility(label):
        index = ['Aaj', 'Kal', 'Parson'].index(label)
        lines[index].set_visible(not lines[index].get_visible())
        plt.draw()

    check_buttons.on_clicked(toggle_visibility)

    # Display the figure
    plt.show()
else:
    print("Error: Empty response for one or more days.")
```

```python
import matplotlib.pyplot as plt
from matplotlib.widgets import CheckButtons
import datetime
from datetime import timedelta

# Sample temperature data
today_temp = 25
yesterday_temp = 22
before_yesterday_temp = 20

# Create a list of timestamps
today = datetime.datetime.now()
yesterday = today - timedelta(days=1)
before_yesterday = today - timedelta(days=2)
timestamps = [today, yesterday, before_yesterday]

# Create a figure and axes
fig, ax = plt.subplots()

# Plot the temperature data
line, = ax.plot(timestamps, [today_temp, yesterday_temp, before_yesterday_temp], label="Temperature", marker='o')

# Set the axes labels
ax.set_xlabel("Time")
ax.set_ylabel("Temperature (°C)")

# Add checkboxes
check_buttons = CheckButtons(ax, ['Today', 'Yesterday', 'Before Yesterday'], [True, True, True])

# Add a callback function to update the graph when a checkbox is checked
def update_graph(label):
    if label == "Today":
        line.set_ydata([today_temp] + [None] * 2)
    elif label == "Yesterday":
        line.set_ydata([None, yesterday_temp] + [None])
    elif label == "Before Yesterday":
        line.set_ydata([None] * 2 + [before_yesterday_temp])
    ax.legend()
    plt.draw()

# Add the callback function to the checkboxes
check_buttons.on_clicked(update_graph)

# Show the graph
plt.show()
```
